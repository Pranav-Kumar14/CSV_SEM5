```python
# eg LOG TRANSFORMATION
img = cv.imread('C:/Users/Shweta Ganesh/Downloads/breast_digital_Xray.tif')
c = 255/(np.log(1+np.max(img)))
log_trans = c* np.log(1+img)
log_trans = np.array(log_trans, dtype=np.uint8)
cv.imshow("transformed_img",log_trans)
cv.waitKey(0)
```

```python
# eg gamma transform
img = cv.imread('C:/Users/Shweta Ganesh/Downloads/breast_digital_Xray.tif')
for gamma in [0.1,0.5,1.2,2.2]:
    gamma_corr = np.array(255*(img/255)**gamma,dtype='uint8')
    cv.imshow(f'gamma_trans{gamma}',gamma_corr)
    cv.waitKey(0)
```

```python
#eg piecewise linear transform 
img = cv.imread('C:/Users/Shweta Ganesh/Downloads/breast_digital_Xray.tif')
def pixelVal(pix,r1,s1,r2,s2):
    if(0<=pix and pix<=r1):
        return (s1/r1)*pix
    elif (r1<pix and pix<=r2):
        return ((s2-s1)/(r2-r1))*(pix-r1)+s1
    else:
        return ((255-s2)/(255-r2)) * (pix-r2) + s2
r1 = 70
s1 = 0
r2 = 140
s2 = 255
pixVal_vec = np.vectorize(pixelVal)
cont_stretch = pixVal_vec(img,r1,s1,r2,s2)
cv.imshow('stretchedd',cont_stretch)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# que 1 histogram equalisation
# parameters - path, 0 (0 is for greyscale)
img = cv.imread('C:/Users/Shweta Ganesh/Downloads/breast_digital_Xray.tif',0)
#parameters = array, no of bins , range of bins last one not inc
hist, bins = np.histogram(img.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_norm = (cdf - cdf.min()) * 255 / (cdf.max()-cdf.min())
cdf_final = cdf_norm.astype('uint8')

eq_img = cdf_final[img]
cv.imshow('og',img)
cv.imshow('eq',eq_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#q2 histogram specification
input_img = cv.imread('C:/Users/Shweta Ganesh/Downloads/hist_match_original_image.tif')
ref_img = cv.imread('C:/Users/Shweta Ganesh/Downloads/hist_match_ref_image.tif')

# if ref_img.dtype != np.uint8:
#     ref_img = cv.normalize(ref_img, None, 0, 255, cv.NORM_MINMAX).astype(np.uint8)

# if input_img.dtype != np.uint8:
#     input_img = cv.normalize(input_img, None, 0, 255, cv.NORM_MINMAX).astype(np.uint8)


input_hist, bins = np.histogram(input_img.flatten(),256,[0,256])
ref_hist, bins = np.histogram(ref_img.flatten(),256,[0,256])

input_cdf = input_hist.cumsum()/ input_hist.sum()
ref_cdf = ref_hist.cumsum()/ref_hist.sum()

mapping = np.zeros(256,dtype=np.uint8)
for i in range(256):
    mapping[i] = np.argmin(np.abs(ref_cdf-input_cdf[i]))

spec_img = mapping[input_img]
cv.imshow('ip',input_img)
cv.imshow('ref',ref_img)
cv.imshow('spec',spec_img)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#q3 resize and cropping
resized_half = cv.resize(input_img, None, fx=0.5, fy=0.5)
resized_exact = cv.resize(input_img, (300,200))
cropped = input_img[50:250,100:400]
cv.imshow("Resized", resized_half)
cv.imshow("Resized2", resized_exact)
cv.imshow("Cropped", cropped)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# eg gaussian blur, median blur and bilateral blur
img = cv.imread('C:/Users/Shweta Ganesh/Pictures/Saved Pictures/HD-wallpaper-steam-workshop-studio-ghibli-studio-ghibli-beautiful.jpg')
cv.imshow("og",img)

gaussian = cv.GaussianBlur(img, (7,7),0)
cv.imshow("gauss",gaussian)

median = cv.medianBlur(img, 5)
cv.imshow("median", median)

bilateral = cv.bilateralFilter(img, 9, 75,75)
cv.imshow("bi",bilateral)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# identity kernal
kernel1 = np.array(
    [
        [0,0,0],
        [0,1,0],
        [0,0,0]
    ]
)

identity = cv.filter2D(src = img, ddepth = -1, kernel = kernel1)
cv.imshow('identity', identity)

# blurring kernel 
kernel2 = np.ones((5,5), np.float32)/25
img2 = cv.filter2D(src=img, ddepth=-1,kernel=kernel2)
cv.imshow('kernel',img2)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# q1 unsharp masking 
# enhance edges by subtracting blurred version from img and adding it to the img again
img = cv.imread('C:/Users/Shweta Ganesh/Downloads/hist_match_ref_image.tif')
gauss = cv.GaussianBlur(img, (5,5), 0)
mask = cv.subtract(img,gauss)
new = cv.add(img, mask)
cv.imshow("og",img)
cv.imshow("unsharp",new)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# 1 create binary images using thresholding 
img = cv.imread('C:/Users/Shweta Ganesh/Downloads/breast_digital_Xray.tif', cv.IMREAD_GRAYSCALE)
i, binary1 = cv.threshold(img,127,255,cv.THRESH_BINARY)
i,binary2 = cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
adaptive = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11,2)
cv.imshow("Original1", img)
cv.imshow("Binary1", binary1)
cv.imshow("Binary Inverse1", binary2)
cv.imshow("Adaptive1", adaptive)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# 2 hough transform 
img = cv.imread('C:/Users/Shweta Ganesh/Downloads/line_detection_2.png')
gray2 = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
edges = cv.Canny(gray2,50,150)
lines = cv.HoughLinesP(edges,1,np.pi/180,100,minLineLength=100,maxLineGap=10)
for line in lines:
    x1,y1,x2,y2 = line[0]
    cv.line(img,(x1,y1),(x2,y2),(0,0,255),2)
cv.imshow('h',img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#harris
img = cv.imread('C:/Users/Shweta Ganesh/Downloads/corner3.jpg')
grey = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
grey = np.float32(grey)
harris = cv.cornerHarris(grey,2,3,0.04)
harris = cv.dilate(harris, None)
img[harris > 0.01 *harris.max()] = [0,0,255]
cv.imshow("Harris Corners1", img)
cv.waitKey(0)
cv.destroyAllWindows()
```
